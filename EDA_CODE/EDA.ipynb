```python
import pandas as pd
import plotly.graph_objects as go

# Define the path to your Excel files
path = r'C:\Users\HP\Downloads\proj\individual\CLEANED_MERGED_RESULT_.xlsx'

# Load the combined DataFrame from the result Excel file
df = pd.read_excel(path)

# Define the columns you want to analyze
numeric_columns = [
    "box1WagesTipsAndOtherCompensations", "box2FederalIncomeTaxWithheld", 
    "box3SocialSecurityWages", "box4SocialSecurityTaxWithheld", 
    "box16StateWagesTips", "box17StateIncomeTax"
]

# Select the columns you want to analyze
selected_columns = df[numeric_columns]

# Calculate the correlation matrix
correlation_matrix = selected_columns.corr()

# Create a heatmap using Plotly
fig = go.Figure(data=go.Heatmap(
    z=correlation_matrix.values,
    x=correlation_matrix.columns,
    y=correlation_matrix.index,
    colorscale='Viridis',
    colorbar=dict(title='Correlation Coefficient'),
    text=correlation_matrix.values,  # Add correlation values as text
    hovertemplate='Correlation: %{text:.2f}'  # Format the hover text
))

# Update layout
fig.update_layout(title='Correlation Matrix Heatmap', xaxis_title='Columns', yaxis_title='Columns')

# Save the heatmap to an HTML file
output_path = r'C:\Users\HP\Downloads\proj\Learning\correlation_matrix_heatmap.html'
fig.write_html(output_path)

print(f"Heatmap of the correlation matrix saved to {output_path}")
```

## KDE

```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
from scipy import stats

# Define the path to your Excel files
path = r'C:\Users\HP\Downloads\proj\individual\CLEANED_MERGED_RESULT_.xlsx'

# Load the combined DataFrame from the result Excel file
df = pd.read_excel(path)

# Define the columns you want to analyze
numeric_columns = [
    "box1WagesTipsAndOtherCompensations", "box2FederalIncomeTaxWithheld", 
    "box3SocialSecurityWages", "box4SocialSecurityTaxWithheld", 
    "box16StateWagesTips", "box17StateIncomeTax"
]

# Create a list to hold all Plotly figure objects
figs = []

# Generate KDE plots for each numeric column
for col in numeric_columns:
    # Drop NaN values for the KDE
    data = df[col].dropna()
    
    # Compute the KDE with a custom bandwidth
    kde = stats.gaussian_kde(data, bw_method='scott')  # 'scott' or 'silverman' for bandwidth
    x = np.linspace(data.min(), data.max(), 1000)  # More points for a smoother curve
    y = kde(x)

    # Create a plotly figure
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='KDE', line=dict(width=2)))
    fig.update_layout(title=f'KDE Plot of {col}', xaxis_title=col, yaxis_title='Density')

    figs.append(fig)

# Save all figures to the specified path
output_path = r'C:\Users\HP\Downloads\proj\Learning\kde_plots.html'

# Create a single HTML file with all the plots
with open(output_path, "w") as f:
    f.write('<html><head></head><body>')
    for fig in figs:
        f.write(fig.to_html(full_html=False, include_plotlyjs='cdn'))
    f.write('</body></html>')

print(f"KDE plots saved to {output_path}")
```
